```python
# Determination of the error in U0/t0 due to the 
# systematic uncertainty in the lattice depth

from uncertainties import ufloat,unumpy
from scipy import stats
from scipy.interpolate import interp1d


wF = np.loadtxt('dataplots/EDAT/tANDU.dat')
wF = wF[ np.logical_and(  wF[ :, 0] > 6.5 , wF[:,0]<7.5  ) ] 
z = np.polyfit( wF[:,0], np.power( wF[:,2], 1./3.) , 1)  
p = np.poly1d(z) 
#
# get the slope of wF^(1/3)  in units of wF^(1/3) 
slope_wF13 = p[0] / p(7.) 
#print slope_wF13  
#x = np.linspace( 6., 8., 20) 
#plt.plot( wF[:,0], np.power(wF[:,2], 1./3.), '.', color='black')
#plt.plot( x, p(x), '-' )  
#plt.show()

# get the uncertainty in the lattice depth for x,y,z
sx = ufloat( 21.8, 0.3)
sy = ufloat( 21.6, 0.2)
sz = ufloat( 18.6, 0.5)

def get_err_sq( s ):
    mW = 7.0 * s.nominal_value 
    return  ( (mW / sx).std_dev )**2.
 
s0_err =  np.sqrt(  get_err_sq( sx ) \
             + get_err_sq( sy ) + get_err_sq( sz ) )

U_err_factor = slope_wF13 * s0_err 

#print slope_wF13
#print get_err_sq( sx ) 
#print get_err_sq( sy ) 
#print get_err_sq( sz ) 
#print  s0_err 
print U_err_factor

import glob 
from colorChooser import rgb_to_hex, cmapCycle


# Start matplotlib
from matplotlib import rc
rc('text', usetex=True)
plt.rcParams['text.latex.preamble'] = [
       r'\usepackage{bm}',        # for bold math
       r'\usepackage{siunitx}',   # i need upright \micro symbols, but you need...
       r'\sisetup{detect-all}',   # ...this to force siunitx to actually use your fonts
       r'\usepackage[textlf]{MyriadPro}',
       #r'\usepackage{helvet}',    # set the normal font here
       r'\usepackage{sansmath}',  # load up the sansmath so that math -> helvet
       r'\sansmath',               # <- tricky! -- gotta actually tell tex to use!
       #r'\renewcommand{\familydefault}{\sfdefault}'
]    
plt.rcParams['axes.linewidth'] = 0.65
plt.rcParams['patch.linewidth'] = 0.5

fig = plt.figure(figsize=(3.50, 2.8))
gs = matplotlib.gridspec.GridSpec( 1,1, wspace=0.25, hspace=0.11,\
                    top=0.865, left=0.13, right=0.80, bottom=0.14)
ax = fig.add_subplot( gs[0] )  

def fx(x):
    return  x * 11.1 / 290. 

    

    
# YLABELS 
for tick in ax.get_yaxis().get_major_ticks():
    tick.set_pad(2.5)
#ax.set_ylabel( r'$\mathrm{Spin\ structure\ factor}\ S_{\mathbf{Q}}$', rotation=90, fontsize=7, labelpad=5. )
ax.set_ylabel( r'$S_{\mathbf{Q}}$', rotation=0, fontsize=10., labelpad=10. )
ax.set_xlabel( r'$U_{0}/t_{0}$', fontsize=10., labelpad=5.) 
    
# AXES GENERAL PROPERTIES 
#ax.grid(alpha=0.3)
ax.axhline( 1.0, ls='-', color='0.7', lw=1.2, zorder=0.1)
ax.set_xlim(2.0, 23.)
ax.set_ylim(0.55, 2.25 ) 


ax.yaxis.set_major_locator( \
    matplotlib.ticker.MultipleLocator( 0.4 ))
ax.yaxis.set_minor_locator( \
    matplotlib.ticker.MultipleLocator( 0.2 ))
ax.xaxis.set_major_locator( \
    matplotlib.ticker.MaxNLocator( nbins=6))


# Setup scattering length ticks on the top
def ascatt(x):
    return x * 290. / 11.1
def Ut(x):
    return x * 11.1 / 290. 
aSs =  [ 80., 200., 290., 380., 470, 560. ] 
ax2 = ax.twiny()
ax2.set_xticks( [ Ut(xi) for xi in  aSs ] )  
ax2.set_xticklabels( [ r'$%d$' % xi for xi in aSs ] )
ax2.tick_params(axis='x', labelsize=7)
ax2.set_xlim( 2. , 23. )
#ax2.set_xlabel('$\mathrm{scattering\ length}\ (a_{0})$', fontsize=7, labelpad=3.5) 
ax2.set_xlabel('$a_{s}\ (a_{0})$', fontsize=10., labelpad=6.5) 

for tick in ax2.get_xaxis().get_major_ticks():
    tick.set_pad(1.1)



#### PLOT THE QMC COMPARISON INCLUDING LDA 

import glob
from colorChooser import rgb_to_hex, cmapCycle

aSs = [200, 290, 380, 470]
Tsets = []
for aS in aSs:
    fdats  = sorted(glob.glob('dataplots/THERM/{:03d}/*.dat'.format(int(aS))))
    Tspis  = [ float(f.split('Tspi')[1].split('.dat')[0]) for f in fdats ] 
    print aS
    print Tspis
    Tsets.append( set(Tspis))
    
    
Tcommon = sorted(set.intersection(*Tsets))
print "TCommon = ", Tcommon

Tcommon = [ 0.53, 0.55, 0.57, 0.61, 0.68, 0.80, 1.09, 1.99]

m  =  ['D','s','^','o','v','p', '<', '>' ]
ms =  [3.6, 3.8, 4.1, 3.8, 4.1, 4.3, 4.1, 4.1]
ms = [ 0.85*msi for msi in ms ]


cs = [ cmapCycle(matplotlib.cm.jet, 0.15 ),\
       cmapCycle(matplotlib.cm.jet, 0.77 ),\
       cmapCycle(matplotlib.cm.spectral, 0.45 ),\
       cmapCycle(matplotlib.cm.jet, 0.86 ),\
       cmapCycle(matplotlib.cm.hsv, 0.8 ),\
       cmapCycle(matplotlib.cm.jet, 0.6 ),\
       cmapCycle(matplotlib.cm.jet, 0.2 ),\
       cmapCycle(matplotlib.cm.jet, 0.2 ),\
       ]

cs = [ cmapCycle(matplotlib.cm.hsv, 0.00, lbound=0, ubound=1.20 ),\
       cmapCycle(matplotlib.cm.hsv, 0.14, lbound=0, ubound=1.20 ),\
       cmapCycle(matplotlib.cm.hsv, 0.28, lbound=0, ubound=1.20 ),\
       cmapCycle(matplotlib.cm.hsv, 0.42, lbound=0, ubound=1.20 ),\
       cmapCycle(matplotlib.cm.hsv, 0.56, lbound=0, ubound=1.20 ),\
       cmapCycle(matplotlib.cm.hsv, 0.70, lbound=0, ubound=1.20 ),\
       cmapCycle(matplotlib.cm.hsv, 0.84, lbound=0, ubound=1.20 ),\
       cmapCycle(matplotlib.cm.hsv, 0.98, lbound=0, ubound=1.20 ),\
       ]

cs = [ cmapCycle(matplotlib.cm.hsv, 1.00, lbound=0, ubound=1.00 ),\
      cmapCycle(matplotlib.cm.autumn, 0.6, lbound=0, ubound=1.0 ),\
       cmapCycle(matplotlib.cm.spring, 1.0, lbound=0, ubound=1.0 ),\
       'blue',\
       cmapCycle(matplotlib.cm.hsv, 0.56, lbound=0, ubound=1.20 ),\
       cmapCycle(matplotlib.cm.hsv, 0.70, lbound=0, ubound=1.20 ),\
       cmapCycle(matplotlib.cm.hsv, 0.84, lbound=0, ubound=1.20 ),\
       cmapCycle(matplotlib.cm.hsv, 0.98, lbound=0, ubound=1.20 ),\
       ]




cs = [ cmapCycle(matplotlib.cm.hsv, 0.00, lbound=0, ubound=1.20 ),\
       cmapCycle(matplotlib.cm.hsv, 0.14, lbound=0, ubound=1.20 ),\
       cmapCycle(matplotlib.cm.hsv, 0.28, lbound=0, ubound=1.20 ),\
       cmapCycle(matplotlib.cm.hsv, 0.42, lbound=0, ubound=1.20 ),\
       cmapCycle(matplotlib.cm.hsv, 0.56, lbound=0, ubound=1.20 ),\
       cmapCycle(matplotlib.cm.hsv, 0.70, lbound=0, ubound=1.20 ),\
       cmapCycle(matplotlib.cm.hsv, 0.84, lbound=0, ubound=1.20 ),\
       cmapCycle(matplotlib.cm.hsv, 0.98, lbound=0, ubound=1.20 ),\
       ]

cs = [  cmapCycle( matplotlib.cm.jet, float(i), lbound=0.,  ubound=7., inv=False) for i in range(8) ]

cs = ['#e41a1c',\
      '#377eb8',\
'#4daf4a',\
'#984ea3',\
'#ff7f00',\
'dodgerblue',\
'#a65628',\
'#f781bf' ]

cs = ['blue'] * 8 

cs = [ cmapCycle(matplotlib.cm.hsv, 0.00, lbound=0, ubound=1.20 ),\
       cmapCycle(matplotlib.cm.hsv, 0.14, lbound=0, ubound=1.20 ),\
       cmapCycle(matplotlib.cm.spectral, 0.55, lbound=0, ubound=1.20 ),\
       cmapCycle(matplotlib.cm.hsv, 0.28, lbound=0, ubound=1.20 ),\
       #cmapCycle(matplotlib.cm.hsv, 0.56, lbound=0, ubound=1.20 ),\
       #'#006666',\
       'steelblue',\
       cmapCycle(matplotlib.cm.hsv, 0.70, lbound=0, ubound=1.20 ),\
       
       cmapCycle(matplotlib.cm.hsv, 0.84, lbound=0, ubound=1.20 ),\
       
       cmapCycle(matplotlib.cm.hsv, 0.98, lbound=0, ubound=1.20 ),\
       ]

resultsThermometer = [] 
Tstars = [] 

import os
rawdir = 'dataplots/THERM/04_rawdat/'
if not os.path.exists(rawdir):
        os.makedirs(rawdir)
rawctr = 0

for i,Tspi in enumerate(Tcommon):
    
    color = cs[i] 
      
    if Tspi == 0.50:
        continue
    result = []
    # Columns are :  N/1e5,  SpiBulk, Frac, S/N_Bulk, t*, T*, U*
    for j,(aS,U,bestN) in enumerate([\
            ( 80,3,0), \
            (120,4,0), \
            (160,6,0), \
            (200,7,0), \
            (245,9,0), \
            (290,11,0), \
            (335,12,0), \
            (380,14,0), \
            (425,16,0), \
            (470,17,0), \
            (515,19,0), \
            (560,21,0), \
            ]):
        try:
            fname = 'dataplots/THERM/{a:03d}/aS{a:03d}_U{U:02d}_Tspi{T:0.2f}.dat'.format(a=aS,U=U,T=Tspi)
            dat = np.loadtxt( fname )
            dat = np.atleast_2d( dat ) 
            # 0 is number
            # 1 is Spi Bulk
            # 3 is Entropy Bulk
            # 4 is t^{*}
            # 5 is T/t^{*} 
            # 6 is U^{*} / t^{*}
            try:
                sth = dat[bestN][7]
            except:
                sth = np.nan
                
            result.append( [ fx(aS), dat[bestN][1], dat[bestN][0], dat[bestN][3] , dat[bestN][5], dat[bestN][6], sth ] ) 
            
            print "U0/t0 = {:0.2f}  T/t0 = {:0.2f},  T/t* = {:0.2f}".format( U, Tspi, dat[bestN][5])
            
            if aS == 290:
                resultsThermometer.append( [  dat[bestN][5], dat[bestN][1] ]  )
        except Exception as e:
            print "Did not find aS={:03d}, T={:0.2f}".format(aS,Tspi)
            print e
            pass
    result = np.array(result)
    print "U/t star  =",  result[:,5]
    #print "mean Tstar = ", np.mean( result[:,4] ) 
    #print " std Tstar = ", np.std( result[:,4] )
    #print result 
    
    # Choose what goes on the legend title
    Tlabel = np.mean(result[:,4] )
    Tstars.append(Tlabel) 
    print "TSTARS:"
    print result[:,4]
    #~Tlabel = Tspi

    lwidth = 0.85
    # Dummy plot for label 
    ax.plot( result[:,0], 5.0*np.ones_like(result[:,0]), #'-', \
            m[i%len(m)]+'-', \
            ms=ms[i], mew=0.25, lw=lwidth, color=color, alpha=1.0,\
               mfc='white', mec=color, label='${:0.2f}$'.format( np.round(Tlabel,2)) , zorder=0.9)
    #SPI
    ax.plot( result[:,0], result[:,1], #'-', \
            m[i%len(m)], \
            ms=ms[i], mew=0.25, lw=0.6, color=color, alpha=1.0,\
               mfc='white', mec=color, zorder=0.9)
    
    flda = interp1d( result[:,0], result[:,1], kind='cubic')
    xnew = np.linspace( result[:,0].min(), result[:,0].max(), 120 )
    ax.plot( xnew, flda(xnew), '-',  lw=lwidth, alpha=0.8, color=color, zorder=0.5)
    
    np.savetxt( rawdir + '{:02d}_T{:0.2f}.dat'.format(rawctr,np.round(Tlabel,2)), \
               np.column_stack((result[:,0], result[:,1])), fmt='%.4g',\
               header='U0/t0  Spi')
    rawctr = rawctr  + 1 
    
        
    #STH
    if np.around(Tlabel,2) in [1.68, 0.70, 0.47]:
        
        
        x = result[:,0]
        y = result[:,6] 
        subset = ~np.isnan(y)
        
        if Tlabel < 0.68:
            subset = ~np.isnan(y) & (x<16.)
        
        ax.plot( x[subset], y[subset], #'-', \
                m[i%len(m)], \
                ms=ms[i], mew=0.25, lw=0.6, color=color, alpha=1.0,\
                   mfc='white', mec=color, zorder=0.9)
        np.savetxt( rawdir + '{:02d}_T{:0.2f}.dat'.format(rawctr,np.round(Tlabel,2)), \
                   np.column_stack((result[:,0], result[:,1])), fmt='%.4g',\
                   header='U0/t0  Sth')
        rawctr = rawctr  + 1 
        try:
            flda = interp1d( x[subset], y[subset], kind='cubic')
            xnew = np.linspace( x[subset].min(), x[subset].max(), 120 )
            ax.plot( xnew, flda(xnew), '-', dashes=[1.,1.], lw=lwidth*0.8, alpha=1.0, color=color, zorder=0.5)
        except:
            pass
    
    
    


figLDA = plt.figure(figsize=(3.4, 3.0))
gsLDA = matplotlib.gridspec.GridSpec(1,1,\
            wspace=0.1, hspace=0.34,\
            left=0.17, right=0.97, bottom=0.16, top=0.97)
axTH = figLDA.add_subplot( gsLDA[0] )

axTH.set_xlabel(r'$T/t_{*}$', fontsize=12., labelpad=6.)
#axTH.set_ylabel(r'\mathrm{Spin\ structure\ factor}\ $S_{\bm{\pi}}$', rotation=90., labelpad=4., fontsize=7)
axTH.set_ylabel(r'$S_{\bm{\pi}}$', rotation=0., labelpad=15., fontsize=12.)
axTH.axhspan( 1.962-0.115, 1.962+0.115, facecolor='gray', alpha=0.6)
resultsThermometer = np.array( resultsThermometer )

x = resultsThermometer[:,0]
x = np.array(Tstars)
#print Tstars
y = resultsThermometer[:,1]
axTH.plot( x, y, 'D', mfc='white', mec='black', ms=4.4, mew=1.2, color='black')
axTH.set_xlim(0.43, 0.74)
axTH.set_ylim(1.3, 2.33)
#axTH.grid(alpha=0.3)

axTH.xaxis.set_major_locator( matplotlib.ticker.MultipleLocator(0.1) )
axTH.xaxis.set_minor_locator( matplotlib.ticker.MultipleLocator(0.05) )

axTH.yaxis.set_major_locator( matplotlib.ticker.MultipleLocator(0.2) )
axTH.yaxis.set_minor_locator( matplotlib.ticker.MultipleLocator(0.1) )





#for tick in axTH.get_xaxis().get_major_ticks():
#    tick.set_pad(2.5)
for tick in axTH.get_yaxis().get_major_ticks():
    tick.set_pad(1.2)
    
    
axTH.text(0.13, 0.08, '$U_{0}/t_{0}=11.1$',\
          ha='left', va='bottom', transform=axTH.transAxes, fontsize=12., linespacing=1.5) 
          #bbox={'boxstyle':'square', 'facecolor':'white'})
    



handles, labels = ax.get_legend_handles_labels()
l0 = legend = ax.legend(handles, labels, title='$T/t_{*}$',bbox_to_anchor=(1.02,0.92), loc='upper left', numpoints=1, \
    prop={'size':9.}, handlelength=1.1, handletextpad=0.4, ncol=1, columnspacing=0.7, labelspacing=0.55)
plt.setp( legend.get_title(), fontsize=9.)
ax.add_artist(l0)

######################################################################

def plotexp( ax, dat, mfc, lstr, **kwargs):
    
    cols = kwargs.pop('cols',(0,3,4))
    
    X=fx(dat[:,cols[0]]); Y=dat[:,cols[1]];YERR=dat[:,cols[2]]
    # In the compressibility paper we found  +/-5% in the lattice depth 
    # is +/-14% in U/t 
    XERR = X * U_err_factor / 2. 
    XERR = X * 0.14
    XERR = X * U_err_factor
    XERR = X * 0.
    marker = kwargs.pop('marker','o')
    mec=kwargs.pop('mec','black')
    
    f2B = interp1d(X, Y-YERR, kind='cubic')
    f2T = interp1d(X, Y+YERR, kind='cubic')
    xnew = np.linspace(X.min(), X.max(),120)

    glw = kwargs.pop('guide_lw',3.)
    gcl = kwargs.pop('guide_color','gray')
    gal = kwargs.pop('guide_alpha', 0.4)
    if kwargs.pop('guide', False):
        ax.fill_between( xnew, f2B(xnew), f2T(xnew), lw=0., color=gcl, alpha=gal, zorder=0)
    
    ax.errorbar( X, Y, yerr=YERR, xerr=XERR, \
            fmt='.', \
            #capsize=0.6, elinewidth=0.4, \
            capsize=0., elinewidth=0.8, \
            mew=0.9, marker=marker,\
            mfc=mfc, mec=mec, ms=4.2, ecolor=mec, **kwargs )
    
    ax.plot( X, Y,  '.', marker=marker, ms=4.2, mew=0.9, mfc=mfc, mec=mec, label=lstr, **kwargs)



datSpi = np.loadtxt('dataplots/EDAT/experiment_data_Spi.txt')
datSth = np.loadtxt('dataplots/EDAT/experiment_data_Sth.txt')

plotexp( ax, datSpi, '0.65', r'$\mathbf{Q}=\bm{\pi}$', mec='black',marker='o',\
        guide_lw=12., guide_color='0.65', guide_alpha=0.6, guide=False)
plotexp( ax, datSth, 'white', r'$\mathbf{Q}=\bm{\theta}$', mec='black', marker='o',\
        guide_lw=4., guide_color='0.68', guide_alpha=0.4, guide=False)

#datSpi = np.loadtxt('dataplots/EDAT/sdata_A2.dat')
#datSth = np.loadtxt('dataplots/EDAT/sdata_A1.dat')
#plotexp( ax, datSpi, '0.65', r'$\mathbf{Q}=\bm{\pi}$', mec='black',marker='s',\
#        guide_lw=12., guide_color='0.65', guide_alpha=0.6, guide=False, cols=(0,1,2))
#plotexp( ax, datSth, 'white', r'$\mathbf{Q}=\bm{\theta}$', mec='black', marker='s',\
#        guide_lw=4., guide_color='0.68', guide_alpha=0.4, guide=False, cols=(0,1,2))

#for dat,mfc,lstr in [(datSpi,'black',r'$\mathbf{Q}=\bm{\pi}$' ), (datSth,'white', r'$\mathbf{Q}=\bm{\theta}$' )]:
#    plotexp( ax, dat, mfc, lstr)
    
handles, labels = ax.get_legend_handles_labels()
legend = ax.legend( handles[-2:], labels[-2:], bbox_to_anchor=(1.,1.), loc='upper right',numpoints=1,prop={'size':9.},\
    handlelength=0.6,handletextpad=0.5, labelspacing=0.26)




ax.tick_params(axis='both', which='major', labelsize=10., length=2.)
ax2.tick_params(axis='both', which='major', labelsize=10., length=2.)

axTH.tick_params(axis='both', which='major', labelsize=10.,  length=1.5)
axTH.tick_params(axis='both', which='minor', labelsize=10.,  length=1.0)


    
import datetime
datestamp = datetime.datetime.now().strftime('%Y%m%d_%H%M%S_')
datestamp = ''

#fig.savefig('dataplots/SEXP/' + datestamp+'udat_final_onerow_XU.png', dpi=350)
fig.savefig('dataplots/THERM/04_SpiLDA_partA.png', dpi=400)
figLDA.savefig('dataplots/THERM/04_SpiLDA_partB.png', dpi=400)
```
