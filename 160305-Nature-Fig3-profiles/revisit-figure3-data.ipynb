```python
import pickle
import numpy as np

gr = 3.700
savedir = '../dataplots/PROF/'
for ii, (aS, mu) in enumerate([(200,0.107), (290,0.079), (380,0.029), (470,-0.027)]):    
    fname = savedir + 'aS{:03d}_gr{:0.3f}_mu{:0.3f}.pck'.format(aS,gr,mu)
    calculation_data = pickle.load( open( fname,"rb" ) )
    
    # Will go ahead and save the calculation data in plain text:
    
    LEN1 = 50
    
    with open('all_data_{:d}a0.dat'.format(aS), 'w') as fout:
        
        fout.write("{} : {}\n".format(
                "Scattering length (a0)".ljust(LEN1), 
                calculation_data['aS']))
        
        fout.write("{} : {}\n".format(
                "Atom number".ljust(LEN1), 
                calculation_data['Number']))
        
        fout.write("{} : {}\n".format(
                "Bulk S_pi (this is the exp. observable)".ljust(LEN1), 
                calculation_data['SpiBulk']))
        
        fout.write("{} : {}\n".format(
                "Temperature in dens. profile calc".ljust(LEN1),
                calculation_data['Tdens']))
        
        fout.write("{} : {}\n".format(
                "Temperature in s_pi profile calc".ljust(LEN1),
                calculation_data['Tspi']))
        
        
        # There was an error in the calculation of the trap-averaged entropy,
        # was getting S/N**2 rather than S/N.  That error is corrected here
        fout.write("{} : {}\n".format(
                "Bulk (i.e. trap-averaged) entropy (kb/N)".ljust(LEN1),
                calculation_data['entrbulk'] * calculation_data['Number']/1e5))
        
        fout.write("\nPROFILES:\n")
        
        fout.write("\n{} : ".format(
                "r111 (evaluation points along 111)".ljust(LEN1)))
        np.savetxt(fout, calculation_data['r111'].ravel(), fmt='%.6g', newline=', ')
        
        fout.write("\n\n{} : ".format(
                "t111 (tunneling along 111)".ljust(LEN1)))
        np.savetxt(fout, calculation_data['t111'].ravel(), fmt='%.6g', newline=', ')
        
        fout.write("\n\n{} : ".format(
                "U111 (onsite inter along 111)".ljust(LEN1)))
        np.savetxt(fout, calculation_data['U111'].ravel(), fmt='%.6g', newline=', ')
        
        fout.write("\n\n{} : ".format(
                "mu111_Er (chem potential along 111)".ljust(LEN1)))
        np.savetxt(fout, calculation_data['mu111_Er'].ravel(), fmt='%.6g', newline=', ')
        
        fout.write("\n\n{} : ".format(
                "n111 (density along 111, calc at T = Tdens)".ljust(LEN1)))
        np.savetxt(fout, calculation_data['n111'].ravel(), fmt='%.6g', newline=', ')
        
        fout.write("\n\n{} : ".format(
                "dens111 (density along 111, calc at T = Tspi)".ljust(LEN1)))
        np.savetxt(fout, calculation_data['dens111'].ravel(), fmt='%.6g', newline=', ')
        
        fout.write("\n\n{} : ".format(
                "entr111 (entropy along 111)".ljust(LEN1)))
        np.savetxt(fout, calculation_data['entr111'].ravel(), fmt='%.6g', newline=', ')
        
        fout.write("\n\n{} : ".format(
                "spi111 (local s_pi along 111)".ljust(LEN1)))
        np.savetxt(fout, calculation_data['spi111'].ravel(), fmt='%.6g', newline=', ')
        
        
        fout.write("\n\n\nPROFILES WITH A FINER GRID:\n")
        
        fout.write("\n{} : ".format(
                "r111_fine (evaluation points along 111)".ljust(LEN1)))
        np.savetxt(fout, calculation_data['r111_fine'].ravel(), fmt='%.6g', newline=', ')
        
        fout.write("\n\n{} : ".format(
                "n111_fine (density along 111, calc at T = Tdens)".ljust(LEN1)))
        np.savetxt(fout, calculation_data['n111'].ravel(), fmt='%.6g', newline=', ')
        
        fout.write("\n\n{} : ".format(
                "spi111_fine (local s_pi along 111)".ljust(LEN1)))
        np.savetxt(fout, calculation_data['spi111_fine'].ravel(), fmt='%.6g', newline=', ')
        
        
        fout.write('\n\n')
    
    
```
